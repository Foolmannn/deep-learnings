```python
import torch
```

# Clearing gradient

```python
x=torch.rand((1),requires_grad=True)
x
```

```python
y=x**2
y
```

```python
y.backward()

x.grad
# tensor([0.2056])
# when we run another pass 
# tensor([0.4112])
```

So when we run the multiple pass the gradience doesnot clears itself . so it will add up to the previous gradience .which is not desirable so we need to clear the gradient before re running the new pass 

```python
x.grad.zero_() # this will clear the gradience for next epoch 
```

# Disable Gradient Tracking 

### Sometimes we donot need to calculate the gradient . For ex when the NN is trained and we want to use it for prediction. During the prediction we donot need the gradient tracking as we just neet the forward pass. As backward pass willl not run anyway/. T his is done because for the big NN this gradient tracking comsumes the memory. so to preserve that we will stop this tracking during the use for predictions 

```python
x= torch.tensor(2.0,requires_grad=True)
x
```

```python
y=x**2
y
```

```python
y.backward()
```

```python
x.grad
```

### so for ex here the training ends . So now we need to stop tracking 

### option 1 - requires_grad_(False)
### option 2 - detach()
### option 3 - torch.no_grad()

## Method 1 : 

```python
x.requires_grad_(False)
```

```python
x # now we donot see the tracking at op 
```

```python
y = x**2 
y
```

```python
y.backward() # this will give error as tracking is stopped. 
```

## Method 2:

```python
x= torch.tensor(2.0,requires_grad=True)
x
```

```python
z=x.detach()
z
```

```python
y = x** 2
y
```

```python
y1= z**2 
y1
```

```python
y.backward()
```

```python
y1.backward() # this will ntot work as z is detached from the x 
```

## Method 3 : 

```python
x= torch.tensor(2.0,requires_grad=True)
x
```

```python
with torch.no_grad():
    y = x ** 2

# y = x ** 2 # with this the backtracking is on 
# now y is not tracking 
```

```python
y.backward()
```
