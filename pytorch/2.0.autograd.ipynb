Why autograd?

```python
#for derivative/differentiation of y=x^2
def dy_dx(x):
    return 2*x

dy_dx(3)
```

```python
# for z=sin(y) y=x^2
#dervateing is 2xcos(x^2)
import math
def dz_dx(x):
    return 2*x*math.cos(x**2)
dz_dx(4)
```

So derivative of the nested function requires the chaining . So in ml we need to calculate the derivates for the multiple reason like loss calculation. Gradient descent. etc. Thus a tool is needed for differentiation. As the NN get complex. coding , hand doing the derivate get difficult. So autograd performs this for us on tensors 

Autograd is a core component of Pytorch that provides automatic differentiation for tensor operations. It enables the gradient computation , which is essential for trqaining machine learning models using optimization algorithms like gradient descent . 

```python
# y=x^2
# y=x^2 , z=sin(y)
# 
```

```python
import torch

x = torch.tensor(3.0,requires_grad=True) # this requires_grad is by default false. this tell pytorch thta this will use the derivative. so x will be usable for differentiantion

```

```python
y = x**2  
print(x,y)  # y when printed shows . tensor(9., grad_fn=<PowBackward0>) this means pytorch. creates computation graph. ie y is derived frm the x . So pytorch keep strack of the function/mathermatic used. 
# here powbackward means pytorch knows that y has been made from x with the power function. This is very useful for the differentiation 
```

```python
y.backward() # this backward does Computes the gradient of current tensor wrt graph leaves.

# The graph is differentiated using the chain rule
```

```python
x.grad # this shows the value of the gradients 
```

```python
x=torch.tensor(4.0,requires_grad=True)
```

```python
y=x**2
```

```python
z=torch.sin(y)
```

```python
print(x,y,z) # we can see the computation graph created. So we can see sinbackward ie. z from the sinof y  
```

```python
# for dz/dx : dz/dy * dy/dx so we need to calcuate the chain rule. So compution graph save the operations made so that it can backtrack and perform derivate and apply chain rule to give us the derivate

z.backward()
```

```python
x.grad  # so as we can see how easily autograd calculates the derivate / gradient 
```

```python
y.grad # The .grad attribute of a Tensor that is not a leaf Tensor is being accessed 

# so in graph we have the leaves, roots and intermediate. So we can apply the backward and gradient on the leaf and root only . not in intermediate 
```
