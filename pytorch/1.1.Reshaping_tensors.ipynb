```python
b = torch.rand(2,3,5)
b
```

```python
# permutations 

b.permute(2,0,1) # this rearrange as 2 to first, 0 to middle and 1 at the end 

# this creates b.reshape(5,3,2)
```

```python
b.reshape(5,3,2) # as we can see it is same as permute one . 
```
