# Using the builtin forward pass . using the activation and linear layers  

So instead of using the handwritten loss functions, activation function. Now we will learn to use the builtin features of the pytorch 

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

### train test split 

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

### Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

### Label Encoding 

```python
encoder = LabelEncoder()  # as diagonisis are in the alphabet we encode them in number her 0 and 1 
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

Numpy arrays to PyTorch tensors

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32)) # this is neccessary as builtin function will use the float instead of the double type tensors
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

### Defining the model

```python
# now using the nn module 
class mySimpleNN(nn.Module):

    def __init__(self,num_features):

        super().__init__()
        self.linear  = nn.Linear(num_features,1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self,features):

        out = self.linear(features)
        out = self.sigmoid(out)
        return out

    def loss_function(self, y_pred, y):
        # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

        # Calculate loss
        loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()  # this is formula for binary cross entropy loss function 
        return loss
    
    
         
```

### Important Parameters

```python
learning_rate = 0.1 
epochs = 25
```

### Training Pipeline 

```python
# create model 
model = mySimpleNN(X_train_tensor.shape[1])

#define loop

for epoch in range(epochs):
    #forward pass 
    y_pred = model(X_train_tensor)
    # print(y_pred)
        

    # loss calculate

    loss = model.loss_function(y_pred,y_train_tensor)
    # print(f"Epoch : {epoch} Loss : {loss.item()}")

    # backward pass
    loss.backward()

    # parameters update  # wnew = wold - lr.( dL / dW)  and new bias  = bold - lr (dL / db) here lr is learning rate and it is called n eta
    with torch.no_grad():    
        model.linear.weight -= learning_rate * model.linear.weight.grad
        model.linear.bias -= learning_rate * model.linear.bias.grad


    # zero gradients as gradient accumulates so clering them

    model.linear.weight.grad.zero_()
    model.linear.bias.grad.zero_()
    

    # print los in each epoch : 
    print(f"Epoch : {epoch} , Loss : {loss.item()}")



# model.weights
```

### Evaluation

```python
# model evaluation 

with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred> 0.7).float()  # if the y_pred is above 0.5 set to 1 
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f'Accuracy : {accuracy.item()}')

print(type(y_pred))
# if y_pred>0.5:
#     y_pred=1
# else:
#     y_pred=0
```
