```python
#create model class
import torch
import torch.nn as nn
# we are implementing simple model with 1 neuron 5 weights and 1 bias . It is binary classification problem 
# we are improving using hidden layer of relu then applying the sigmoid 
class Model(nn.Module):
    def __init__(self, num_features):

        super().__init__()
        self.linear1 = nn.Linear(num_features,3)  # first layer
        self.relu = nn.ReLU()  #hidden using relu
        self.linear2 = nn.Linear(3,1)  # takes three weights/inputs gives 1 using sigmoid activation function 
        self.sigmoid = nn.Sigmoid()

    def forward(self,features):

        out = self.linear1(features)
        out=self.relu(out)
        out=self.linear2(out)
        out= self.sigmoid(out)

        return out
    
```

```python
#create dataset 

features = torch.rand(10,5)

# create model
model = Model(features.shape[1])

# call model for forward pass

# model.forward(features) pytorch has implemented autocall of the forward method . so if we call object as the function the forward of the class auto runs 

model(features)
```

```python
model.linear1.weight # we get 15 weights 
```

```python
model.linear2.weight # we get 3 weights 
```
