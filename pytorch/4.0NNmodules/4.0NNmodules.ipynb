```python
#create model class
import torch
import torch.nn as nn
# we are implementing simple model with 1 neuron 5 weights and 1 bias . It is binary classification problem 
class Model(nn.Module):
    def __init__(self, num_features):

        super().__init__()
        self.linear = nn.Linear(num_features,1 ) 
        self.sigmoid = nn.Sigmoid()

    def forward(self,features):

        out = self.linear(features)
        out= self.sigmoid(out)

        return out
    


```

```python
#create dataset 

features = torch.rand(10,5)

# create model
model = Model(features.shape[1])

# call model for forward pass

# model.forward(features) pytorch has implemented autocall of the forward method . so if we call object as the function the forward of the class auto runs 

model(features)
```

```python
# show model weights 

model.linear.weight  # we get 5 wieghts 
```
