Step 1: Split Your Data¶
Use the train_test_split function to split up your data.

Give it the argument random_state=1 so the check functions know what to expect when verifying your code.

Recall, your features are loaded in the DataFrame X and your target is loaded in y.

````python
# import thge train_test_split
from sklearn.model_selection import train_test_split

# split the csv data into two set 
# here train_X are the features values to be used in training and train_y are also the target values used in training
# And val_X , val_y are going to used for the validation and they will not be used in training   

train_X,val_X,train_y,val_y = train_test_split(X,y,random_state=1) # this sequence matters
````

`train_test_split()` is a function from `sklearn.model_selection` used to split data into **training** and **testing/validation** sets.

Your code:

```python
train_X, val_X, train_y, val_y = train_test_split(
    X,
    y,
    random_state=1
)
```

means:

* `X` → features (inputs)
* `y` → target labels (outputs)
* `train_X` → training features
* `val_X` → validation/testing features
* `train_y` → training targets
* `val_y` → validation/testing targets

Suppose your dataset is:

```python
X = [
    [1, 10],
    [2, 20],
    [3, 30],
    [4, 40],
    [5, 50]
]

y = [100, 200, 300, 400, 500]
```

Each row in `X` corresponds to one value in `y`:

| X      | y   |
| ------ | --- |
| [1,10] | 100 |
| [2,20] | 200 |
| [3,30] | 300 |
| [4,40] | 400 |
| [5,50] | 500 |

---

### What `train_test_split()` does internally

By default:

```python
train_test_split(X, y)
```

uses:

```python
test_size = 0.25
```

So:

* 75% → training
* 25% → testing/validation

For 100 rows:

| Set        | Rows |
| ---------- | ---- |
| train      | 75   |
| validation | 25   |

---

### How rows are chosen

Before splitting, sklearn **shuffles** data randomly.

Example:

Original order:

```text
0  1  2  3  4  5  6  7
```

After shuffle:

```text
5  1  7  2  4  0  3  6
```

Then split:

Training:

```text
5 1 7 2 4 0
```

Validation:

```text
3 6
```

So actual rows change each time unless:

```python
random_state=1
```

is set.

---

### What `random_state=1` means

This fixes the randomness.

Without:

```python
train_test_split(X,y)
```

each run gives different splits.

With:

```python
train_test_split(X,y,random_state=1)
```

same rows go to train/test every run.

Think:

```text
random_state = seed for random number generator
```

---

### How to know exactly which rows went where

You can print them:

```python

print("Validation X:")

```

and:

```python
print("Training y:")

print("Validation y:")

```

### To know original row indices

Use dataframe indices:

Example:

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.DataFrame({
    'A':[1,2,3,4,5],
    'target':[10,20,30,40,50]
})

X = df[['A']]
y = df['target']

train_X, val_X, train_y, val_y = train_test_split(
    X,
    y,
    random_state=1
)

print(train_X.index)
print(val_X.index)
```

Output might be:

```text
Index([4,0,3], dtype='int64')   # train rows
Index([2,1], dtype='int64')     # validation rows
```

Meaning:

Original rows:

```text
Train → rows 4,0,3
Validation → rows 2,1
```

```python
    random_state=1

```

```python

```

30% test:

```python

```
